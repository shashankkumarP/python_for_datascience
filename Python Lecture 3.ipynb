# WHILE LOOP

```python
x = 0

while x < 10:
    print('x is currently: ',x)
    print('x is still less than 10, adding 1 to x')
    x+=1 # x = x+ 1 updation
```

```python
x = 0

while x < 10:
    print('x is currently: ',x)
    print('x is still less than 10, adding 1 to x')
    x+=1 
    if x == 4:
        print('Breaking because x == 4')
        break    
```

```python
x = 0

while x < 10:
    print('x is currently: ',x)
    print('x is still less than 10, adding 1 to x')
    x+=1 
    if x == 4:
        print('Breaking because x == 4')
        break
    else:
        print('Continuing')
                
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue  #Skips its 
  print(i)
```
